### 01 - Ingest CMS Medicare Part D data from Azure Blob Storage

This notebook replaces the direct CMS download and instead pulls data from your **Azure Blob Storage (Gen2)**.

**Prerequisites:**
1. Azure Storage Account with Hierarchical Namespace enabled.
2. Container named `testingminjdemo`.
3. Folder structure `raw_data/` containing `2023.csv`.

```python
# Parameters - Update these or use Fabric Environment Variables
storage_account_name = "YOUR_STORAGE_ACCOUNT_NAME"
container_name = "testingminjdemo"
relative_path = "raw_data/2023.csv"
lakehouse_dir = "Files/cms_raw"

mssparkutils.fs.mkdirs(lakehouse_dir)

# Construct ABFSS path
source_path = f"abfss://{container_name}@{storage_account_name}.dfs.core.windows.net/{relative_path}"
destination_path = f"{lakehouse_dir}/2023.csv"

print(f"Copying from {source_path} to {destination_path}...")

try:
    mssparkutils.fs.cp(source_path, destination_path)
    print("Data ingestion complete!")
except Exception as e:
    print(f"Error copying data: {str(e)}")
    print("Ensure your Fabric Workspace has the necessary permissions (Storage Blob Data Reader) on the Azure Storage Account.")
```
